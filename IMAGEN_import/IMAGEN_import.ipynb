```python
import pandas as pd
import numpy as np
import glob, os
from operator import itemgetter

data_path='/nfs/zorba/DOWNLOADED/IMAGEN/MPRAGE_FACES_REST_REWARD/BL/processed/nifti'
# Below will capture all the structural and functional dicom folders (some are duplicates)
T1_glob_data = '*/SessionA/ADNI_MPRAGE/*.nii.gz' # Grabs all baseline, later can add the repeat session
faces_glob_data = '*/SessionA/EPI_faces/*.nii.gz' # Come back and incorporate
#reward_glob_data = '*/SessionB/EPI_reward/*.nii.gz'
rest_glob_data = '*/SessionB/EPI_rest/*.nii.gz'

os.chdir(data_path)
T1_scans=sorted(glob.glob(T1_glob_data)) # Focus on T1 scans for now
faces_scans=sorted(glob.glob(faces_glob_data))
#reward_scans=sorted(glob.glob(reward_glob_data))
rest_scans=sorted(glob.glob(rest_glob_data))

print "length of structural scans is: " + str(len(T1_scans))
print "length of faces scans is: " + str(len(faces_scans))
#print "length of reward scans is: " + str(len(reward_scans))
print "length of rest scans is: " + str(len(rest_scans))

# Now get only unique ids for structural scans
T1_sub_ids=[]
for s in T1_scans:
    T1_sub_ids.append(s.split('/')[0])

# Now get only unique ids for functional lores resting scans
faces_sub_ids=[]
for s in faces_scans:
    faces_sub_ids.append(s.split('/')[0])

# # Now get only unique ids for functional lores resting scans
# reward_sub_ids=[]
# for s in reward_scans:
#     reward_sub_ids.append(s.split('/')[0])

# Now get only unique ids for functional lores resting scans
rest_sub_ids=[]
for s in rest_scans:
    rest_sub_ids.append(s.split('/')[0])
    
gen_file = '/nfs/zorba/DOWNLOADED/IMAGEN/TABLES/gender_hand_center.csv'
df = pd.read_csv(gen_file,delimiter=';',dtype=str)
df = df.set_index('code_in_study')
id_list = df.index.tolist()

# convert to string
#id_list = list(map(str,id_list))
#print id_list
#print type(T1_sub_ids[1])

age_file = '/nfs/zorba/DOWNLOADED/IMAGEN/TABLES/age_at_mri_acquisition_BL.csv'
#age_file = '/nfs/zorba/DOWNLOADED/IMAGEN/TABLES/PSYTOOLS-neo-youth_BL_2017-07-15_08_51_42.csv'
df_age = pd.read_csv(age_file,delimiter=',',dtype=str)
df_age = df_age.set_index('PSC2')
id_list_age = df_age.index.tolist()

subject_dict = {}
count=0

for i in range(len(T1_sub_ids)):
    item=T1_sub_ids[i] # Item is the subject ID "i.e. A00053278"
    T1_path=T1_scans[i]
    #print (item)
    #print file_path
    if item in id_list:
        # Here go ahead and grab the subject's functional paths if they exist
       
        try:
            idx = faces_sub_ids.index(item)
            faces_path  = faces_scans[idx]
        except: 
            faces_path = []
        
#         try:
#             idx = reward_sub_ids.index(item)
#             reward_path  = reward_scans[idx]
#         except: 
#             reward_path = []
        
        try:
            idx = rest_sub_ids.index(item)
            rest_path  = rest_scans[idx]
        except: 
            rest_path = []    
        
        #print id_list.index(item)
        count += 1
        df_index =id_list.index(item)
        sex = df.iloc[df_index]['gender']
        site  = df.iloc[df_index]['name']
        
        try:
            df_age_index = id_list_age.index(item)
            age_days = df_age.iloc[df_age_index]['ADNI MPRAGE']
            age = int(age_days)/365.25
        except:
            age="unknown"
            
        # make sure age and sex are defined
        if sex=="unknown" or age=="unknown":
            pass
        else:
            subject_dict[item] = {'age': "", "sex": "", "T1_path":"", "faces_path":"","rest_path":"", "site":""}
            subject_dict[item]['age'] = age
            subject_dict[item]['sex'] = sex
            subject_dict[item]['site'] = site
            subject_dict[item]['T1_path'] = T1_path
            subject_dict[item]['faces_path'] = faces_path
            #subject_dict[item]['reward_path'] = reward_path
            subject_dict[item]['rest_path'] = rest_path
       
print (count, len(T1_sub_ids))
print len(subject_dict)

for key, value in subject_dict.items()[0:5]:
    print key  
    print value["age"]
    print value["sex"]
    print value["site"]
    print value["T1_path"]
    print value["faces_path"]
    print value["rest_path"]
        
```
